```python
#Now to split data to carry out linear regression study
```

```python
#Now need to split into train/test
from sklearn.model_selection import train_test_split #split data from sklearn
from sklearn.linear_model import LinearRegression #import linear regression
```

```python
train_test_split
```

```python
customers.columns
```

```python
#Setting X = memebership length
X = customers[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
#Note: need to omit string value columns!
```

```python
#Setting y = amount spent
y = customers['Yearly Amount Spent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
lm = LinearRegression()#generate Linear Regression object
```

```python
lm
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)#shows intercept
print(lm.coef_)#shows coeff for each feauture - relates to X_train columns (i.e features)
```

```python
X_train.columns
```

```python
coefs = pd.DataFrame(lm.coef_,X.columns,columns=['Coeffs'])
```

```python
coefs
```

```python
predictions = lm.predict(X_test)#This is set that model has not seen before!
```

```python
predictions
```

```python
plt.scatter(y_test,predictions)#Plot of test targets and predictions
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
sns.distplot(y_test-predictions, bins=50)#Now make a dist plot of the residuals
```

```python

#Regression evaluation metrics
from sklearn import metrics
metrics.mean_absolute_error(y_test, predictions)

np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

```python
metrics.mean_squared_error(y_test, predictions)
```
