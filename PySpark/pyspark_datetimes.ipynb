```python
# Always neds to be done in Rasberry Pi
import findspark
findspark.init('/home/baxman/spark-2.4.7-bin-hadoop2.7')
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('dates').getOrCreate()
```

```python
df = spark.read.csv('/home/baxman/Codes/PySpark/Python-and-Spark-for-Big-Data-master/Spark_DataFrames/appl_stock.csv',
                    header = True, inferSchema = True)
```

```python
df.head(1)
```

```python
df.select('Date', 'Close').show()
```

```python
from pyspark.sql.functions import dayofmonth, hour, year, dayofyear, month, weekofyear,format_number, date_format
```

```python
df.select(dayofmonth(df['Date'])).show()
```

```python
df.select(weekofyear(df['Date'])).show()
```

```python
df.select(hour(df['Date'])).show()
```

```python
df.select(month(df['Date'])).show()
```

```python
df.select(year(df['Date'])).show()
```

```python
# Making new column for the year
df.withColumn("YEAR",year(df['Date'])).show()
```

```python
# Making new df

new_df = df.withColumn("YEAR",year(df['Date']))
```

```python
# Groupng by

new_df.groupBy("Year").mean().select(['YEAR','avg(Close)']).show()
```

```python
result = new_df.groupBy("Year").mean().select(['YEAR','avg(Close)'])
```

```python
result.withColumnRenamed('avg(Close)', 'Av per Year').show()
```

```python
new_result = result.withColumnRenamed('avg(Close)', 'Av per Year')
```

```python
new_result.select(['YEAR',format_number('Av per Year',2)]).show()
```
