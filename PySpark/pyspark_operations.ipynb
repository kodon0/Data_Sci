```python
# Always neds to be done in Rasberry Pi
import findspark
findspark.init('/home/baxman/spark-2.4.7-bin-hadoop2.7')
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Operations').getOrCreate()
```

```python
df = spark.read.csv('/home/baxman/Codes/PySpark/Python-and-Spark-for-Big-Data-master/Spark_DataFrames/appl_stock.csv',
                   inferSchema = True, header=True)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
# Filtering using SQL

df.filter("Close < 500").select('open').show()
```

```python
# Filter using pyspark

df.filter(df['Close'] < 500).select('open').show()
```

```python
# Filtering based on multiple conditions - need to split up conditions into each parenthesis

df.filter((df['Close'] < 200) & (df['Open'] > 200)).show()
```

```python
# Grab a specific row - 'what date had the close price is 197'

df.filter(df['Low']==197.16).show()
```

```python
# In SQL

df.filter("Low = 197.16").show()
```

```python
# Use collect instead of show, so you can save it into a variable

result = df.filter(df['Low']==197.16).collect()
```
